```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_locacao_imoveis.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['dados_locacao'])
```

```python
dados.head()
```

```python
colunas = list(dados.columns)
```

```python
dados = dados.explode(colunas[1:])
```

```python
dados.reset_index(inplace = True, drop = True)
```

```python
dados.info()
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace(' reais', '').replace(',', '.').strip())
```

```python
import numpy as np
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].astype(np.float64)
```

```python
dados
```

```python
dados['apartamento'] = dados['apartamento'].str.replace('\(blocoAP\)', ' ', regex = True).str.strip()
```

```python
dados['datas_combinadas_pagamento'] = pd.to_datetime(dados['datas_combinadas_pagamento'], dayfirst=True)
```

```python
dados['datas_de_pagamento'] = pd.to_datetime(dados['datas_de_pagamento'], dayfirst=True)
dados.head(1)
```

```python
dados['atraso'] = (dados['datas_de_pagamento'] - dados['datas_combinadas_pagamento']).dt.days
```

```python
media_atraso = dados.groupby(['apartamento'])['atraso'].mean()
```
